```python

# Parameter LCG
a = 13 # Pengali
m = 516 # Modulus
Z0 = 18 # Seed (Z0)
n = 10 # Jumlah nilai yang akan dihasilkan
```

* a = 13 (Pengali): Nilai yang digunakan untuk mengalikan nilai sebelumnya dalam rumus LCG. Mempengaruhi pola angka acak.

* m = 516 (Modulus): Menentukan batasan nilai hasil, angka yang dihasilkan akan berada dalam rentang 0 hingga 515.

* Z0 = 18 (Seed): Nilai awal yang digunakan untuk memulai perhitungan angka acak pertama. Nilai ini sangat mempengaruhi urutan angka acak yang dihasilkan.

* n = 10 (Jumlah nilai): Jumlah angka acak yang akan dihitung menggunakan rumus LCG.

Rumus LCG yang digunakan adalah:

    Zi = (a.Zi-1) mod m

Dimana Zi adalah angka acak yang dihitung pada iterasi ke-i, dan

    Ui = Zi/m

adalah angka acak seragam dalam rentang (0, 1).



```python
# Daftar username peserta
username_peserta = [
    "Inara", "Rusdi", "Hanif", "Malika", "Sarah",
    "Isna", "Saifudin", "Rama", "Angel", "Salwa",
    "Putri", "Diana", "Sinta", "Raffa", "Husdan",
    "Syafiq", "Triana", "Windi","Baila", "Laura",
    "Kirana", "Akmal", "Yoan", "Viaz", "Syakib",
    "Namita","Amanda", "Nuca", "Melinda", "Afifah"]
```

* username_peserta: List yang berisi nama-nama peserta dalam bentuk string.

* Daftar ini bisa digunakan untuk mengakses, memilih, atau melakukan operasi lainnya pada setiap username, misalnya untuk memilih pemenang dari peserta-peserta yang terdaftar.

```python
def lcg():  # Mendefinisikan fungsi untuk algoritma LCG
    hasil = []
    z = Z0  # Set nilai awal (Z0)
    for _ in range(n):
        z = (a * z) % m  # Rumus LCG
        u = round(z / m, 3)  # Perhitungan Ui dan pembulatan ke 3 desimal
        hasil.append((z, u))
    return hasil  # Mengembalikan hasil
```

1. hasil = [ ]: Membuat list kosong untuk menyimpan hasil perhitungan.

2. z = Z0: Menginisialisasi nilai awal (seed) untuk variabel z dengan nilai Z0.

3. for _ in range(n): Melakukan perulangan sebanyak n kali, sesuai dengan jumlah nilai yang ingin dihasilkan.

4. z = (a * z) % m: Menghitung nilai baru z dengan rumus LCG: (Zi=a.Zi-1)mod m.

5. u = round(z / m, 3): Menghitung angka acak seragam Ui dengan membagi z dengan m, kemudian membulatkannya ke 3 desimal.

6. hasil.append((z, u)): Menyimpan pasangan nilai (z, u) ke dalam list hasil.

7. return hasil: Mengembalikan list hasil yang berisi angka acak yang telah dihitung.

```python
# Menghasilkan angka LCG
hasil = lcg()  # Memanggil fungsi 'lcg' yang telah didefinisikan
```

Fungsi ini akan mengembalikan hasil berupa daftar angka acak yang dihitung, yang kemudian disimpan dalam variabel hasil.


* hasil = lcg(): Memanggil fungsi lcg() untuk menghasilkan dan menyimpan angka acak dalam variabel hasil.

```python
# Menampilkan hasil dalam format tabel
print(f"{'i':<3} | {'Zi-1':<15} | {'Zi':<15} | {'Ui':<10}")
print("-" * 50)
for i, (z, u) in enumerate(hasil, start=1):
    Z_prev = hasil[i-1][0] if i > 1 else "24"  # Menampilkan Zi-1 untuk baris pertama menggunakan "24"
    print(f"{i:<3} | {Z_prev:<15} | {z:<15} | {u:<10}")
```

1. **print(f"{'i':<3} | {'Zi-1':<15} | {'Zi':<15} | {'Ui':<10}")** :
Menampilkan header tabel dengan kolom i, Zi-1 (nilai sebelumnya), Zi (nilai yang dihitung), dan Ui (angka acak seragam).

2. **print("-" * 50)** :
Menampilkan garis pemisah untuk memisahkan header dari data yang akan ditampilkan.

3. **for i, (z, u) in enumerate(hasil, start=1)** :
Melakukan perulangan pada setiap elemen dalam list hasil, yang berisi pasangan nilai z (angka acak) dan u (angka acak seragam), dimulai dari indeks 1.

4. **Z_prev = hasil[i-1][0] if i > 1 else "24"** :
Menentukan nilai Zi-1 (nilai sebelumnya). Pada iterasi pertama, Z_prev diatur ke "24", jika tidak, diambil dari elemen sebelumnya dalam list hasil.

5. **print(f"{i:<3} | {Z_prev:<15} | {z:<15} | {u:<10}")** :
Menampilkan nilai i, Zi-1, Zi, dan Ui dalam format tabel.

```python
# Simulasi pemenang giveaway berdasarkan Ui
total_peserta = len(username_peserta)  # Jumlah total peserta
jumlah_pemenang = 5  # Jumlah pemenang
```

* **total_peserta = len(username_peserta)** :
Menghitung jumlah total peserta dengan menggunakan fungsi len() pada list username_peserta, yang berisi daftar nama peserta.

* **jumlah_pemenang = 5**:
Menetapkan jumlah pemenang yang diinginkan dalam giveaway, dalam hal ini adalah 5 orang pemenang.

```python
# Memetakan Ui ke username peserta
pemenang = []
for _, u in hasil[:jumlah_pemenang]:
    nomor_peserta = int(u * total_peserta)  # Pemetaan Ui ke nomor peserta
    pemenang.append(username_peserta[nomor_peserta % total_peserta])  # Menggunakan modulo untuk siklus peserta

print("\nPemenang Giveaway:")
print(pemenang)
```

1. **pemenang = [ ]**:
Menyiapkan list kosong untuk menyimpan nama pemenang.

2. **for _, u in hasil[:jumlah_pemenang]**:
Melakukan iterasi pada hasil (nilai yang dihasilkan oleh LCG), dengan mengambil u (nilai acak seragam) dari setiap elemen dalam hasil. Iterasi ini dibatasi hanya untuk jumlah_pemenang yang telah ditentukan.

3. **nomor_peserta = int(u * total_peserta)**:
Memetakan nilai Ui ke nomor peserta dengan mengalikan u dengan total peserta, kemudian membulatkannya menjadi bilangan bulat.

4. **pemenang.append(username_peserta[nomor_peserta % total_peserta])**:
Menggunakan operator modulo (%) untuk memastikan nomor peserta yang dipilih berada dalam rentang indeks username_peserta (sehingga siklus kembali ke awal jika nomor peserta lebih besar dari jumlah peserta). Nama peserta yang terpilih disimpan dalam list pemenang.

5. **print("\nPemenang Giveaway:") dan print(pemenang)**:
Menampilkan daftar pemenang yang terpilih berdasarkan pemetaan Ui.

Secara keseluruhan, kode ini mengonversi nilai acak Ui menjadi nomor peserta dan menentukan pemenang dengan memetakan nilai tersebut ke daftar nama peserta.
